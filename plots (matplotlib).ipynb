```python
# Plot Lat vs Max Humidity
x_values = cities_df['Lat']
y_values = cities_df['Humidity']
plt.scatter(x_values,y_values)
plt.xlabel('Latitud')
plt.ylabel('Humidity')
plt.show()
```
